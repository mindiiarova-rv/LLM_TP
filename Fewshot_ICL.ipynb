````python
import re

zeroshot_prompt = """
You're an expert in sentiment analysis. Your task is to classify the sentence emotion after <<<>>> with one of the following predefined labels:

sadness
joy
love
anger
fear
surprise

You will only respond with the label. Do not include the word "Label". Do not provide explanations or notes.

<<<
Sentence: {sentence}
Label:
>>>
""".strip()


def generate(prompt, llm=llm, generation_config=generation_config):

  # Create turns with the given prompt
  turns = [
    {'role':'user', 'content':prompt}
  ]

  # Tokenize turns.
  input_ids = tokenizer.apply_chat_template(turns, return_tensors='pt').to('cuda')

  # Ensure we don't use gradient to save memory space and computation time.
  with torch.no_grad():
    outputs = llm.generate(
      input_ids,
      generation_config
    )

  # Recover and decode answer.
  answer_tokens = outputs[0, input_ids.shape[1]:-1]
  return tokenizer.decode(answer_tokens).strip()


def parse_answer(answer):
    pattern = r'\b(sadness|joy|love|anger|fear|surprise)\b'
    match = re.search(pattern, answer)

    if match:
        return match.group(0)
    else:
        return ""
````

````python
from tqdm import tqdm

correct_answers = 0
missing_answers = 0
total_samples = len(test)

for example in tqdm(test):
    sentence = example['text']
    true_label = example['label']

    # Generate the prompt
    prompt = zeroshot_prompt.format(sentence=sentence)

    answer = generate(prompt)

    parsed_answer = parse_answer(answer)

    if parsed_answer == "":
        missing_answers += 1
    else:
        if parsed_answer == true_label:
            correct_answers += 1

accuracy = correct_answers / total_samples
missing_ratio = missing_answers / total_samples

print(f"Accuracy: {accuracy:.2f}")
print(f"Ratio of missing answers: {missing_ratio:.2f}")
````

````python
fewshot_prompt = """
You're an expert in sentiment analysis. Your task is to classify the sentence emotion after <<<>>> with one of the following predefined labels:

sadness
joy
love
anger
fear
surprise

You will only respond with the label. Do not include the word "Label". Do not provide explanations or notes.

####
Here are some examples:

{examples}
####

<<<
Sentence: {sentence}
Label:
>>>
""".strip()

def format_demo(demo):
  sentence = demo['text']
  label = demo['label']
  return f"Sentence: {sentence}\nLabel: {label}\n"

def format_demos(demos):
  return "\n".join(format_demo(demo) for demo in demos)

def get_random_demo(k, train=train):
  return random.sample(train, k)
````

## 3. Fewshot Classification - Vector-based Retrieval

Now, we want to improve demonstrayion by the vector representation of our sentence. This is close to what we did when we used RAG on wikipedia page. But here, we'll do it manually and step by step.

To do so, we need to calculate the vector representation of our training dataset. To do this, we'll code a function that returns a vector for a given example. We'll use our LLM hidden states to do this. It's not optimal, but we won't have to load another model.

First, look at the mistral architecture:

There are 36 transformer layers and 1 language model (LM) layer. Each layer will take the following shape: [1, N_TOKENS, N_PARAMS]. We want to extract the vector of the last token from the last transformer. To do so:
- Encode the sentence without any template. `tokenizer.encode(...)`
- Use the `output_hidden_states` keyword of the llm forward function.
- Select the last transformer layer (be careful, don't take the LM layer).
- Select the last token.
- Convert the vector to numpy `.to('cpu').float().numpy()` and return it.

There is a cell below to test your code. The output should be:

 -0.62109375]

````python
def get_hidden_repr(text, llm=llm, prompt_template=zeroshot_prompt):
  input_ids = tokenizer.encode(text, return_tensors='pt').to('cuda')

  with torch.no_grad():
      outputs = llm(input_ids, output_hidden_states=True)

  last_layer_hidden_states = outputs.hidden_states[-1]
  last_token_hidden_states = last_layer_hidden_states[0, -1, :]

  return last_token_hidden_states.to('cpu').float().numpy()
````

````python
from sklearn.metrics.pairwise import cosine_similarity

def compute_similarity(example_a, example_b):
  vector_a = example_a["vector"].reshape(1, -1)
  vector_b = example_b["vector"].reshape(1, -1)

  similarity = cosine_similarity(vector_a, vector_b)

  return similarity[0][0]
````

````python
def get_k_similar_demo(example, k, train=train):
    similarities = [(compute_similarity(example, train_example), train_example) for train_example in train]

    similarities.sort(key=lambda x: x[0], reverse=True)

    top_k_similar_examples = [train_example for _, train_example in similarities[:k]]

    return top_k_similar_examples
````

## 4. Constrained Decoding

Last exercise, we will use the `outlines` package to do constrained generation. This main idea is to guide the generation of the LLM to get the good output formats.

We will use the choices module. Here is the documentation: https://dottxt-ai.github.io/outlines/latest/reference/generation/choices/

There is an example below on how to use it on 1 example. We let you apply this methods to the test dataset. You need report:

- Report them for k=1 and k=5

````python
from outlines import models, generate

def predict_label_with_outlines(example, k, train=train):
    demos = get_k_similar_demo(example, k, train)
    formatted_demos = format_demos(demos)
    prompt = fewshot_prompt.format(examples=formatted_demos, sentence=example["text"])
    choices = ["sadness", "joy", "love", "anger", "fear", "surprise"]

    generator = generate.choice(model, choices)
    generated_text = generator(prompt)

    parsed_answer = generated_text.strip()

    return parsed_answer

def evaluate_model(test, k_values, train=train):
    results = {}

    for k in k_values:
        correct_answers = 0
        missing_answers = 0
        total_samples = len(test)

        for example in tqdm(test, desc=f"Processing k={k}"):
            predicted_label = predict_label_with_outlines(example, k, train)
            true_label = example["label"]

            if predicted_label == "":
                missing_answers += 1
            elif predicted_label == true_label:
                correct_answers += 1

        accuracy = correct_answers / total_samples
        missing_ratio = missing_answers / total_samples
        results[k] = {"accuracy": accuracy, "missing_ratio": missing_ratio}

    return results

k_values = [1, 5]
results = evaluate_model(test, k_values, train)

for k, metrics in results.items():
    print(f"##### k={k} #####")
    print(f"Accuracy:  {metrics['accuracy']:.2f}")
    print(f"Missing Ratio:  {metrics['missing_ratio']:.2f}")
````

**Question: Now that you've used all these solutions, when should you use zeroshot? when should you use fewshot? when should you use constrained decoding?** Zero-shot learning is ideal for simple, general tasks where the model has been pre-trained on diverse data and resources are limited. Few-shot learning is best for task-specific adaptation when we have a smal amount of labeled data and need improved performance. Constrained decoding is suitable for tasks requiring structured or controlled outputs, ensuring consistency and adherence to specific formats.
